# Data Preprocessing

Removing Extra features and maintenance cost as they dont give any meaningful inferences in determining the cluster labels

```python
filtered_data.drop(['Extra Features',"Maintenance cost"],axis=1, inplace=True)
```

```python
filtered_data.info()
```

Removing null values

```python
# Select columns with float datatype
float_columns = filtered_data.select_dtypes(include=['float64','int64']).columns

# Replace null values with mean of each column
filtered_data[float_columns] = filtered_data[float_columns].fillna(filtered_data[float_columns].mean())
```

Transforming string columns to numeric columns 

```python
object_columns=['Used it for', 'Owned for', 'Ridden for']
```

```python
filtered_data = pd.get_dummies(filtered_data, columns=object_columns)
```

```python
filtered_data.dropna(inplace=True)
```

```python
filtered_data.info()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from nltk.stem import PorterStemmer
import re
import gensim
import pandas as pd
import numpy as np

# Sample data

# Step 1: Separate Data by Model Names
model_names = filtered_data["Model Name"].unique()
data_by_model = {model: filtered_data[filtered_data["Model Name"] == model] for model in model_names}

# Step 2: Text Preprocessing
def preprocess_text(text):

    # Lowercase the text
    text = str(text).lower()
    # Remove special characters
    text = re.sub(r'\W', ' ', text)
    # Apply stemming
    ps = PorterStemmer()
    text = ' '.join([ps.stem(word) for word in text.split() if word not in gensim.parsing.preprocessing.STOPWORDS])
    return text

for model, df_model in data_by_model.items():
    df_model["review"] = df_model["review"].dropna().apply(preprocess_text)

# Step 3: Word Embedding using Word2Vec
# Train Word2Vec model
sentences = [review.split() for df_model in data_by_model.values() for review in df_model["review"]]
word2vec_model = gensim.models.Word2Vec(sentences, vector_size=100, window=5, min_count=1, workers=4)

# Step 4: Convert word embeddings to average vectors and update 'review' column
for model, df_model in data_by_model.items():
    average_vectors = []
    for sentence in df_model["review"]:
        word_vectors = []
        for word in sentence.split():
            if word in word2vec_model.wv:
                word_vector = word2vec_model.wv[word]
                word_vectors.append(word_vector)
        if word_vectors:
            average_vector = np.mean(word_vectors, axis=0)
            average_vectors.append(np.array(average_vector))
        else:
            average_vectors.append(np.zeros(100))  # If no word vectors found, use zero vector

    # Step 5: PCA (Reduce to 2 components for visualization)

    review_vectors = np.stack(df_model["review"].to_numpy())


    pca = PCA(n_components=2)
    pca_result = pca.fit_transform(average_vectors)

    print(type(pca_result))
    # Step 6: KMeans Clustering
    kmeans = KMeans(n_clusters=3)
    kmeans.fit(pca_result)
    cluster_labels = kmeans.labels_
    
    # Assign cluster labels back to the original DataFrame
    df_model["cluster_label"] = cluster_labels  
    
    # Print the cluster labels for this model
    print(f"Cluster labels for {model}:")
    print(df_model[["review", "cluster_label"]])  
```

using regex we get only numbers if it’s a price range take avg of both

```python
import re
prices = [
    '₹ 13.99 - 17.4 L',
    '₹ 12.49 - 13.64 L',
    '₹ 17.74 - 19.24 L',
    '₹ 21.99 - 25.88 L',
    '₹ 23.79 - 23.98 L',
    '₹ 1.06 - 1.12 Cr',
    '₹ 1.8 Cr',
    '₹ 29.15 L',
    '₹ 1 Cr',
    '₹ 1.16 Cr',
    '₹ 1.5 Cr',
    '₹ 1.01 - 1.19 Cr'
]

# Regular expression pattern to extract numerical values
pattern = r'(\d+(\.\d+)?)'

# Extract numerical values and calculate average if two numbers are present
numerical_prices = []
for price in prices:
    numerical_values = re.findall(pattern, price)
    if len(numerical_values) == 1:
        numerical_prices.append(float(numerical_values[0][0]))
    elif len(numerical_values) == 2:
        avg_price = (float(numerical_values[0][0]) + float(numerical_values[1][0])) / 2
        numerical_prices.append(avg_price)

print(numerical_prices)
```

```python
data4["PriceRange"]=numerical_prices
```
